### Note

positional argument is followed by keyword argument whenever a function is called 

Incorrect way :

s2=pd.Series(index=["cricket","basketball","polo"],[12,13,14])

Correct way :

s2=pd.Series([12,13,14],index=["cricket","basketball","polo"])

#### Arithmetic operations in series

Syntax -> seriesname.f()

where f() can be sum(),min(),max(),mean(),std(),var(),median()

also sort_values() is a method used to sort the values in increasing order of index in the series

```python
#### example of sorted values in series
s1.sort_values()
```

```python
d2=pd.DataFrame([["a",23,True],["b",21,False]], columns=["Letter","Number","Boolean"])
```

```python
d2
```

```python
d2.index=['first','second']
```

```python
d2
```

```python
d2.rename(index={'first':'1st'})
```

#### but this will again revert back to original index when called again.

```python
d2
```

#### to avoid this, we need to set inplace= True

```python
d2.rename(index={'first':'1st'}, inplace=True)
```

```python
d2
```

```python
d2
```

#### similar concept for column change

```python
d2.rename(columns={'Number':'Numbers'},inplace=True)
```

```python
d2
```

```python
df=d2[:]
```

```python
df.drop('second')
```

```python
df
```

again we need to set inplace as True to prevent the dataframe from reverting back

```python
df.drop('second',inplace=True)
```
