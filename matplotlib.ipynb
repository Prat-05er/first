```python
yp=np.random.randint(10,40,5)
s=np.array(yp)
plt.plot(s,linestyle='dotted',marker="*",mec='pink',ms=20,mfc='b',linewidth=3,color='k')
plt.show()
```

```python
xp=np.random.randn(5)
p=np.array(xp)
yp=np.random.randint(10,40,5)
s=np.array(yp)
plt.plot(p,s,linestyle='dotted',marker="*",mec='pink',ms=20,mfc='b',linewidth=3,color='k')
plt.show()
```

#### Two seperate graphs

```python
#### ADD TITLE USING title() FUCNTION
plt.title("Y vs X")
y1=np.array(np.random.randint(10,100,5))
y2=np.array(np.random.randint(13,46,5))
plt.plot(y1)
plt.plot(y2,color='y')

#### Putting labels in the respective axes

plt.xlabel("X-axis", fontsize=14, color='k',loc='right') ### for location of label we use loc
plt.ylabel("Y-axis", fontsize=14, color='r')

### to add gridlines
plt.grid(color='g', linestyle='dashed')

plt.show()
```
