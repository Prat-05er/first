```python
def rem_out(col):
    sorted(col)
    q1,q3=col.quantile([0.25,0.75])
    iqr=q3-q1
    l_ran=q1-1.5*iqr
    u_ran=q1+1.5*iqr
    return l_ran,u_ran
```
